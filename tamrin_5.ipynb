```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# nltk.download()
```

```python
dataFrame = pd.read_csv("./tamrin 5.csv")
dataFrame.head()
stop_words = set(stopwords.words('english'))
stop_words
```

```python
def abstract_words():
    author_keywords_counts = dict()
    for key, part in enumerate(dataFrame["Abstract"]):
        try:
            keywords = part.split()
            for word in keywords:
                word = word.lower()
                if word[0] == " ":
                    word = word[1:]
                characters = [".", ",", "[", "]", ")", ";", "(", ":", "%", "_", "/", "'"]
                for character in characters:
                    word = word.replace(character, "")
                if word.isdigit() or len(word) == 1:
                    continue
                if word in author_keywords_counts:
                    author_keywords_counts[word] += 1
                else:
                    author_keywords_counts[word] = 1
        except:
            continue
    return dict(sorted(author_keywords_counts.items(), key=lambda kv: kv[1], reverse=True))
```

```python
abstract_words()
```

```python
wordsDataFrame = pd.DataFrame(abstract_words().items(), columns=['words', 'count'])
wordsDataFrame.head()

```
